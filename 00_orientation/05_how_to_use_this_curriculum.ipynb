## How each notebook is structured

Every working session notebook follows the same structure. Once you know the pattern, you can navigate any notebook instantly.

**What you'll explore** — Learning objectives at the top. These tell you exactly what you should be able to do by the end. Read them first. They orient everything that follows.

**How we got here** — A short "previously on..." section connecting this notebook to what came before. If you are confused about a concept, the answer is often in a notebook referenced here.

**Why this matters** — The real-world context for the topic. Why does anyone care about this? Where does it show up? This section gives you the motivation before the mechanics.

**Try it yourself** — The interactive widget section. This is the heart of the notebook. Do not just watch the widget run — interact with it. Every slider, every dropdown, every option exists because it reveals something about the concept.

**What's happening?** — The explanation behind the widget. The math, the intuition, the theory — explained in plain English first, notation second.

**Real-world example** — A chart or mini-analysis showing the concept applied to real data. This connects the abstract to the concrete.

**Key takeaway** — A single bold blockquote sentence. If you remember nothing else from the notebook, remember this.

**Next up** — The final line pointing to the next notebook in the sequence.

## How to use the interactive widgets

The widget is not a demo. It is a tool for building intuition.

**Move every slider to its extremes.** What happens when a parameter is very large? Very small? Zero? Negative? Extreme values often reveal the underlying structure of a concept in ways that moderate values do not.

**Try every dropdown option.** Each option shows you a different view of the same concept. The full picture is only visible when you have tried them all.

**Try to break it.** What happens at nonsensical input values? What happens when you combine settings that should not work together? The best learning often comes from seeing the widget fail in an informative way.

**Form a prediction before you move the slider.** "If I increase σ, the curve should get wider." Then check. This turns passive observation into active hypothesis testing — the core skill in data science.

> **The widget is a tool for building intuition, not a demo to watch. You will get out of it exactly as much active engagement as you put in.**

## These notebooks are reference material

This is probably the most important thing to understand about this curriculum.

**These notebooks are not meant to be mastered in one sitting.** They are designed to be returned to repeatedly — before class, during class, after class, before a lab, and whenever a concept comes up again in a later section and you want to re-check your understanding.

**We do not expect you to memorize this content.** We expect you to know that it exists and how to find it. When you encounter cross-validation in a lab and cannot remember the details, you should reach for `ml_concepts/` rather than starting from scratch.

**Here is how to use each notebook:**

Before class: skim the "What you'll explore" objectives and the "Key takeaway" so you know what the session is about.

During class: follow along with the notebook open. Run cells as the instructor runs them. Note things that confuse you — that is what discussion time is for.

After class: go back and actually interact with the widget on your own, now that you have the context from the session. Read "What's happening?" more carefully.

Before a lab: skim the relevant working session notebook to remind yourself of the key concepts. The lab expects you to apply them — the working session shows you how they work.

**The curriculum is not a race. It is a reference library.**

## How working sessions connect to labs

The curriculum has three types of student-facing content, and they build on each other deliberately.

**Working sessions** are these notebooks. They introduce a concept through explanation, visualization, and interactivity. They are not graded. They are the foundation.

**Labs** are Jupyter notebooks where you apply the concept yourself, on a new dataset, with less scaffolding. Labs have self-grading tests at the bottom that tell you immediately whether your code is correct. They are meant to be done independently, after the working session.

**Assignments** are deeper applications — often combining multiple concepts from different working sessions into a single, open-ended analysis. They require more judgment and less hand-holding than labs.

The rule: do not attempt a lab until you have spent time with the corresponding working session. The lab assumes the working session is part of your preparation. Students who skip the working session and go straight to the lab almost always get stuck in ways that take much longer to resolve than the working session would have taken.

## Tips for independent study

Some practical advice that makes a real difference.

**Read the key takeaway first if you are short on time.** The takeaway is the one thing we most want you to remember. If you only have ten minutes before a class, read the objectives and the takeaway. You will follow the session much better.

**Run every code cell yourself — do not just read it.** Reading code and understanding code are different. The act of running the cell, seeing the output, and tweaking one thing forces engagement that reading cannot.

**If you are stuck, check "How we got here."** The answer to "why is this confusing me?" is often in a previous notebook. The working sessions are designed to build sequentially.

**Write your own notes in markdown cells.** Add a cell after any section that confuses you and write one sentence explaining it in your own words. If you cannot write the sentence, you have not understood it yet. That is the signal.

**Treat confusion as a question, not a failure.** "I don't understand this" is not a statement about your ability. It is a question waiting to be asked. Write it down. Bring it to class. Ask it.

> **Every expert in this room was once a beginner who kept showing up. There is no other path.**
