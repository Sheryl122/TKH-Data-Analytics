## Why Python?

Python won the data science language wars, and it was not a close race. By the mid-2010s, Python had become the dominant language for data work, overtaking R, MATLAB, SAS, and others that had been around for decades. The chart below shows where things stand today.

```python
import plotly.graph_objects as go
from tkh_utils import PALETTE, FONT, base_layout

languages = ["Python", "SQL", "R", "SAS / SPSS", "Julia", "Scala", "MATLAB"]
popularity = [94, 58, 31, 12, 8, 7, 6]
colors = [PALETTE["primary"]] + [PALETTE["muted"]] * (len(languages) - 1)

layout = base_layout(
    title="Language Popularity Among Data Scientists (2024 Survey)",
    xaxis_title="% of Practitioners Using Regularly",
    yaxis_title=""
)
layout.update(
    xaxis=dict(range=[0, 108]),
    yaxis=dict(autorange="reversed"),
    height=360,
    margin=dict(l=120, r=30, t=70, b=60),
    showlegend=False,
)

fig = go.Figure(data=[
    go.Bar(
        x=popularity,
        y=languages,
        orientation="h",
        marker_color=colors,
        text=[f"{p}%" for p in popularity],
        textposition="outside",
    )
], layout=layout)

fig.show()
```

**Why Python specifically?** Three reasons:

Python reads like English. You can often understand what a block of Python code does without knowing anything about programming. That low barrier to entry matters when the goal is doing data science, not becoming a software engineer.

Python has the ecosystem. Libraries like pandas, scikit-learn, and TensorFlow were built by and for data scientists. The community is enormous, which means when you get stuck, the answer is almost always one search away.

Python is versatile. The same language handles data cleaning, statistical analysis, machine learning, web scraping, building APIs, and automating workflows. You do not need to switch languages as your work expands.

## The key libraries

A library is a collection of pre-written code that solves a specific problem. You import a library to use its tools without writing everything from scratch. Here are the ones you will use in this curriculum:

```python
# numpy — the foundation for all numerical computing in Python
# It gives you fast, multidimensional arrays and the math operations to work with them.
import numpy as np
arr = np.array([1, 2, 3, 4, 5])
print("Mean:", np.mean(arr), "| Std:", np.std(arr).round(2))
```

```python
# pandas — the workhorse for data manipulation
# Think of it as a programmable spreadsheet: rows, columns, filtering, grouping, joining.
import pandas as pd
df = pd.DataFrame({"name": ["Alice", "Bob", "Carol"], "score": [88, 74, 92]})
print(df.sort_values("score", ascending=False))
```

```python
# plotly — interactive charts for exploration and communication
# Every chart you build in this curriculum uses Plotly. Charts you can zoom, hover, and explore.
import plotly.express as px
df_temp = px.data.gapminder().query("year == 2007")
fig = px.scatter(df_temp, x="gdpPercap", y="lifeExp", color="continent",
                 size="pop", log_x=True, title="GDP vs Life Expectancy (2007)")
fig.show()
```

```python
# scikit-learn — machine learning in Python
# One consistent API for dozens of algorithms. You will use this constantly.
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
X, y = make_regression(n_samples=100, n_features=1, noise=10, random_state=42)
model = LinearRegression().fit(X, y)
print(f"Learned coefficient: {model.coef_[0]:.2f}")
```

```python
# statsmodels — statistical modeling with detailed output
# Where scikit-learn focuses on prediction, statsmodels focuses on inference: p-values,
# confidence intervals, hypothesis tests — the language of statistics.
import statsmodels.api as sm
import numpy as np
X_sm = sm.add_constant(X)
results = sm.OLS(y, X_sm).fit()
print(results.summary().tables[1])
```

```python
# scipy — scientific computing and statistics
# Statistical tests, probability distributions, optimization, signal processing.
# The Swiss army knife for scientific Python.
from scipy import stats
group_a = [78, 82, 85, 90, 72]
group_b = [88, 91, 95, 87, 93]
t_stat, p_val = stats.ttest_ind(group_a, group_b)
print(f"t-statistic: {t_stat:.3f} | p-value: {p_val:.3f}")
```

## Jupyter notebooks

You are reading this inside a Jupyter notebook. Jupyter is the environment where almost all hands-on data science work happens, at least in early stages of a project.

A notebook is made up of **cells**. Some cells contain markdown text (like this one). Others contain code that you can run. You run a code cell by clicking it and pressing **Shift + Enter**. The output appears immediately below.

This interactivity is what makes Jupyter so useful for data science. You can load a dataset in one cell, plot it in the next, realize something looks odd, go back and fix it, and re-run — all without leaving the document. The notebook becomes a record of your thinking as well as your code.

By the end of this curriculum, you will be comfortable navigating notebooks quickly, organizing your code into logical sections, and using markdown cells to explain your reasoning as you go.

> **A well-organized Jupyter notebook is not just analysis. It is communication. Future you — and your colleagues — will thank present you for clear, readable notebooks.**

## Git and GitHub

Git is version control software. It tracks every change you make to your code, lets you roll back to any previous state, and makes it possible for multiple people to work on the same project without overwriting each other's work.

GitHub is a website that hosts Git repositories. It is where most open-source code in the world lives, including all of the libraries we just discussed.

You do not need to be a Git expert on day one. But you do need to know the basics: how to save (commit) your work, how to look at your history, and how to push your code to GitHub so it is backed up and shareable. We cover this in the tools/ section of the curriculum.

Think of Git as the undo history for your project — except it never forgets, and you can always go back.

## The terminal

The terminal (also called the command line or shell) is a text-based interface for controlling your computer. Instead of clicking through menus, you type commands.

A lot of people find the terminal intimidating at first. This is mostly unfamiliarity. The commands you actually need as a data scientist fit on one page, and once you know them, you will find that the terminal is faster and more precise than any graphical interface for many tasks.

In this curriculum, you will use the terminal to: install packages, run Python scripts, manage your Git repository, and navigate between folders. We cover this in the tools/ section. Do not worry about memorizing anything yet — you will pick it up by using it.

## How the tools fit together

The diagram below shows how all of these tools connect in a typical data science workflow:

```python
import plotly.graph_objects as go
from tkh_utils import PALETTE, FONT, base_layout

# Nodes: label, x, y, color
nodes = [
    ("VS Code", 0.5, 0.9, PALETTE["primary"]),
    ("Jupyter
Notebook", 0.5, 0.62, PALETTE["primary"]),
    ("Python", 0.5, 0.34, PALETTE["primary"]),
    ("NumPy
pandas
scipy", 0.15, 0.08, PALETTE["muted"]),
    ("scikit-learn
statsmodels", 0.5, 0.08, PALETTE["secondary"]),
    ("Plotly", 0.85, 0.08, PALETTE["accent"]),
    ("Git / GitHub", 0.15, 0.62, PALETTE["muted"]),
]

# Edges: (from_idx, to_idx)
edges = [(0,1),(1,2),(2,3),(2,4),(2,5),(0,6)]

edge_traces = []
for (i, j) in edges:
    x0, y0 = nodes[i][1], nodes[i][2]
    x1, y1 = nodes[j][1], nodes[j][2]
    edge_traces.append(go.Scatter(
        x=[x0, x1, None], y=[y0, y1, None],
        mode="lines",
        line=dict(color=PALETTE["muted"], width=1.5),
        showlegend=False,
        hoverinfo="skip",
    ))

node_trace = go.Scatter(
    x=[n[1] for n in nodes],
    y=[n[2] for n in nodes],
    mode="markers+text",
    marker=dict(size=44, color=[n[3] for n in nodes], opacity=0.75,
                line=dict(width=2, color="white")),
    text=[n[0] for n in nodes],
    textposition="middle center",
    textfont=dict(size=9, family=FONT["family"], color="white"),
    showlegend=False,
    hoverinfo="text",
)

layout = base_layout(title="The Data Science Toolkit — How the Tools Connect")
layout.update(
    xaxis=dict(range=[-0.1, 1.1], showgrid=False, zeroline=False, showticklabels=False, visible=False),
    yaxis=dict(range=[-0.05, 1.05], showgrid=False, zeroline=False, showticklabels=False, visible=False),
    height=420,
    margin=dict(l=20, r=20, t=70, b=20),
)

fig = go.Figure(data=edge_traces + [node_trace], layout=layout)
fig.show()
```
